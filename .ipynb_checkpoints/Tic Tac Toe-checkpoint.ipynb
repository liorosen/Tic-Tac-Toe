```python
def Greeding():
    marker = ' '
    marker = input().upper() 
    while  marker != 'X' and  marker != 'O':
            print("You Chose " +  marker)
            print("Please choose again the sign")
            print("----------------------------")
            marker = input("Which sign would u like to be ? X | O").upper()
    print("-------------------------------------")
    
    if  marker == 'X':
        return ('X', 'O')
    elif marker == 'O':
        return ('O', 'X')
   
        
    
        
```

```python
from IPython.display import clear_output
def GameBoard(board):
   
    print("    " + board[1] +" | " +  board[2] + " | " + board[3] + "   ")
    print("   ----------")
    print("    " + board[4] +" | " +  board[5] + " | " + board[6] + "   ")
    print("   ----------")
    print("    " + board[7] +" | " +  board[8] + " | " +  board[9] + "   ")
    
```

```python
def Choosing_Place(board , marker , position):
    board[position] = marker
    
                
```

```python
def Winning(board , marker):
    return ((board[1] == marker and board[2] == marker and board[3] == marker) or # first row
            (board[4] == marker and board[5] == marker and board[6] == marker) or # second row
            (board[7] == marker and board[8] == marker and board[9] == marker) or # third row
            (board[1] == marker and board[4] == marker and board[7] == marker) or # first column
            (board[2] == marker and board[5] == marker and board[8] == marker) or # second column
            (board[3] == marker and board[6] == marker and board[9] == marker) or # third column
            (board[1] == marker and board[5] == marker and board[9] == marker) or # right to left diagnol
            (board[3] == marker and board[5] == marker and board[7] == marker)) # left to right diagnol
```

```python
import random
def choose_first_player():
    if random.randint(0,1) == 0 :
        return 'Player 2 '
    else:
        return 'Player 1 '
```

```python
def space_free(board,position):
    return board[position] == ' '
```

```python
def Board_isFull(board):
    for i in range(1,10):
        if space_free(board, i):
            return False
    return True
```

```python
def Player_Choise(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_free(board,position):
        try:
            position = int(input('Please choose number between 1-9 '))
        except:
            print("Oops this is not a number between 1-9")
            continue
        else:
            break
    return position
```

```python
def Rematch_Game():
    print("Would you like to play again? Y = yes | N = no")
    answer = input().upper()
    if( answer == "N"):
        print("See ya next time")
        return False
    else:
        Greeding()
```

```python
test_board = ['#','X','O','X','O','X','O','X','O','X']
GameBoard(test_board)
```

```python
print("Welcome to the big Tic Tac Toe Game \n")
print("Which sign would u like to be ? X | O") 
game_still_going = True
while True:
    player1Marker , player2Marker = Greeding()
    boardGame = [' '] * 10
    turn = choose_first_player()
    print(turn + ' will be the first to play')
    while game_still_going:
        if turn == 'Player 1':
            print("\n")
            print("Player 1 it's your turn")
            
            position = Player_Choise(boardGame)
            print("Player 1 you Chose " + str(position) + "\n")
            Choosing_Place(boardGame , player1Marker , position)
            GameBoard(boardGame)
            if  Winning(boardGame , player1Marker):
                GameBoard(boardGame)
                print(' Player 1 You are the winner!')
                game_still_going = False
            elif Board_isFull(boardGame):
                GameBoard(boardGame)
                print(' this is a draw!')
                game_still_going = False
            else:
                turn = 'Player 2'
        else:
            print("\n")
            print("Player 2 it's your turn")
            
            position = Player_Choise(boardGame)
            print("Player 2 you Chose " + str(position) + "\n")
            Choosing_Place(boardGame , player2Marker , position)
            GameBoard(boardGame)
            if  Winning(boardGame , player2Marker):
                GameBoard(boardGame)
                print(' Player 2 You are the winner!')
                game_still_going = False
            elif Board_isFull(boardGame):
                GameBoard(boardGame)
                print('this is a draw!')
                game_still_going = False
            else :
                turn = 'Player 1'
    if not Rematch_Game():
        break
```
